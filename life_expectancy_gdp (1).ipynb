```python
import csv
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy

gdp_data = pd.read_csv('all_data.csv')
print(gdp_data.head())
```

```python
gdp_data.describe(include='all')
```

```python
#Creating lists for all the x and y values so I can create a scatter plot for each of the 6 countries seperately.

chile_data = gdp_data[gdp_data['Country'] == 'Chile']
chile_xvalues = chile_data['GDP']
chile_yvalues = chile_data['Life expectancy at birth (years)']

china_data = gdp_data[gdp_data['Country'] == 'China']
china_xvalues = china_data['GDP']
china_yvalues = china_data['Life expectancy at birth (years)']

germany_data = gdp_data[gdp_data['Country'] == 'Germany']
germany_xvalues = germany_data['GDP']
germany_yvalues = germany_data['Life expectancy at birth (years)']

mexico_data = gdp_data[gdp_data['Country'] == 'Mexico']
mexico_xvalues = mexico_data['GDP']
mexico_yvalues = mexico_data['Life expectancy at birth (years)']

usa_data = gdp_data[gdp_data['Country'] == 'United States of America']
usa_xvalues = usa_data['GDP']
usa_yvalues = usa_data['Life expectancy at birth (years)']

zimbabwe_data = gdp_data[gdp_data['Country'] == 'Zimbabwe']
zimbabwe_xvalues = zimbabwe_data['GDP']
zimbabwe_yvalues = zimbabwe_data['Life expectancy at birth (years)']

```

```python
#Now going to create scatter plots across all 6 countries.
    # Create a scatter plot for each country
    
# Create the plot for Chile - GDP vs Life Expectancy
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Life Expectancy vs GDP (scatter plot to show relationship)
ax.scatter(chile_xvalues, chile_yvalues, color='tab:blue', s=100, alpha=0.6, edgecolors='black', linewidth=1.5)

# Add a trend line to show the relationship
import numpy as np
z = np.polyfit(chile_xvalues, chile_yvalues, 1)
p = np.poly1d(z)
ax.plot(chile_xvalues, p(chile_xvalues), color='tab:red', linestyle='--', linewidth=2, label=f'Trend Line')

# Set labels and title
ax.set_xlabel('GDP (Billions USD)', fontsize=12)
ax.set_ylabel('Life Expectancy (years)', fontsize=12)
ax.set_title('Chile: GDP vs Life Expectancy', fontsize=14, fontweight='bold')

# Format x-axis to show GDP in billions
ax.set_xticks([0, 0.5e11, 1e11, 1.5e11, 2e11, 2.5e11, 3e11])
ax.set_xticklabels(['$0B', '$50B', '$100B', '$150B', '$200B', '$250B', '$300B'])

# Add grid and legend
ax.grid(True, alpha=0.3)
ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
```
